### Functions that may reveal useful:

```python
def build_covariates(path, image_index_array):
    """
    INPUT:
     path: path to the images
     image_index_array: an array of image indexes
    OUTPUT:
     an array of dimension (len(image_index_array)) x (38804).
     Each row is a flattened version of the corresponding image.
    """
    covariates = np.zeros((len(image_index_array),38804))
    for i in range(len(image_index_array)):
        img_index = image_index_array[i]
        img = read_image(path_to_img, img_index).reshape(1, 38804)
        covariates[i] = img
    return covariates

def loss(beta, covariates, data):
    """
    INPUT:
     beta: parameter of the logistic regression
     covariates: an array of dimension (minibatch_size) x (38804)
     data: array of dimension (minibatch) containing the gender of each covariate
    OUPUT:
     the loss function when evaluated on the covariates contained in the array "covariates"
    """
    dot_products = np.dot(covariates, beta)
    return np.sum(np.log(1. + np.exp(-data*dot_products)))


#gradient of the loss function with respect to "beta", the first argument of the function "loss"
grad_loss = grad(loss, argnum=0)
```

```python
dim = dim_x * dim_y  
beta_init = np.random.normal(loc = 0, scale = 1. / np.sqrt(dim), size = dim)
```

```python
def probabilistic_predictions(beta, covariate):
    """
    output: list of probabilistic prediction
    proba(y=1|x) = 1/(1+exp(-<x,beta>))
    """
    dot_products = np.dot(covariate, beta)
    proba_vector = 1. / (1. + np.exp(-dot_products))
    return proba_vector

def predictions(beta, covariate):
    """
    output: list of prediction
    """
    proba_vector = probabilistic_predictions(beta,covariate)
    nb_of_examples = covariate.shape[0]
    predictions = np.zeros(nb_of_examples)
    predictions[proba_vector>=0.5] = 1
    predictions[proba_vector<0.5] = -1
    return predictions

def accuracy(beta, covariates, data):
    """
    output: compute the error rate
    """
    prediction_list = predictions(beta, covariates)
    return np.mean(prediction_list == data)

```

### The following function shows our main stochastic gradient descent algorithm. It can be run in different ways with the change of some parameters

#### General Note:

1. The loss function and accuracy are computed based on the test set, since its size is relatively small and representative as well.
2. We will see the effect of change of different parameters on various sample sizes. Then used the best set of parameters to do the SGD on the size 190000 training (the final run)

```python
def sgd(train_img_index, test_img_index, learning_rate = 0.001, nb_batch = 10, threshold = 1, \
        nb_iterations = 1000, selection = "Round-Robin", mom = False, decay = "Step", l = 0.01):
    """
    INPUT:
     train_img_index: indices of the training images
     test_img_index: 
     learning_rate: the initial learning rate
     nb_batch: number of each batch
     threshold: the difference bound of termination
     nb_iterations: Max number of iternations 
     selection: "Round-Robin" or "Random"--The batch selection type
     mom: False or True--Without or With momentum
     decay: "Step" or "1/t"--The way of decay
     l: lambda in 1/t decay
     
    OUTPUT:
     beta: trained beta
     
    Also produced:
     The plot of loss function vs iteration
     The plot of accuarcy vs iteration
     Time per epoch
    """
    # Training and test data
    nb_train = len(train_img_index)
    test_img = build_covariates(path_to_img, test_img_index)
    test_gender = data_gender[test_img_index]
    
    # initialize variables
    loss_function_history = np.zeros(nb_iterations)
    test_accuracy_history = np.zeros(nb_iterations)
    beta = beta_init
    index = 0
    nb_epoch = 0
    start_time = time.time()
    # for the latest epoches, calculate the mean loss
    # record the current mean loss at 1, 2, 3,... epoches
    # record previous mean loss at       2, 3, 4,... epoches  
    prev_mean_loss = 10000000000
    curr_mean_loss = 10000000000
    mean_accuracy = 0
    
    counter = 0
    
    # The batch selection type: Random or Round-Robin
    if (selection == "Round-Robin"):
        select = lambda x: np.arange(x, min((x + nb_batch), nb_train))
    elif (selection == "Random"):
        select = lambda x: random.sample(range(nb_train), nb_batch)

    for k in range(nb_iterations):
        batch_index = train_img_index[select(index)]
        # image of the batch
        batch = build_covariates(path_to_img, batch_index)
        # gender of the batch
        batch_gender = data_gender[batch_index]
        # SGD
        gradient = grad_loss(beta, batch, batch_gender)
        # Adjustment of momentum
        curr_beta = beta
        if (mom == True) and (k>0):
            momentum = curr_beta - prev_beta
        else:
            momentum = 0
        prev_beta = curr_beta
        
        if (decay == "1/t"):
            learning_rate = learning_rate * (1 + l * k) / (1 + l * (k + 1))
        beta = beta - learning_rate * gradient + 0.9 * momentum
        
        loss_function_history[k] = loss(beta, test_img, test_gender)
        test_accuracy_history[k] = accuracy(beta, test_img, test_gender)
        index += nb_batch
        if index >= nb_train: # completed a epoch, reset and add the number of epoched 
            # test
            # print("The number of epoch: ", nb_epoch, "at time: ", time.time() - start_time)
            index = index % nb_train
            nb_epoch += 1
            
            # step decay
            if (nb_epoch > 0) & (nb_epoch % 5 == 0):
                if (decay == "Step"): 
                # reduce the learning rate after every 5 epoches
                    learning_rate /= 2

                # for the latest 5 epoches, calculate the mean loss
                # record the current mean loss at 5, 10, 15,... epoches
                # record previous mean loss at       10, 15, 20,... epoches        
                if (nb_epoch > 5):
                    prev_mean_loss = curr_mean_loss
                curr_mean_loss = np.mean(loss_function_history[((nb_epoch-5) * nb_train // nb_batch):k])
                
                mean_accuracy = np.mean(test_accuracy_history[((nb_epoch-5) * nb_train // nb_batch):k])
                # A termination condition: The error does not reduce significantly
                if (abs(curr_mean_loss - prev_mean_loss) < threshold):
                    break
    end_time = time.time()
    time_per_epoch = (end_time - start_time) / nb_epoch
    
    # loss function plot
    plt.plot(loss_function_history[:k], "-", color = "red", lw=3)
    plt.suptitle("Loss v.s. Iteration of Gradient Descent")
    titlestr = "Training size="+str(nb_train)+",  Test size="+str(len(test_img_index))+\
              ",  Number of batches="+str(nb_batch)+",  " + selection + " version" + \
             ",  Momentum is "+ str(mom) + ",  Decay type is " + decay
    plt.title(titlestr)
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()
    
    # lo
    plt.plot(test_accuracy_history[:k], "-", color = "blue", lw=3)
    plt.suptitle("Test accuracy v.s. Iteration of Gradient Descent")
    plt.title(titlestr)
    plt.ylabel("Accuracy")
    plt.grid(True)
    plt.show()

    print("Time per epoch: {}".format(time_per_epoch))
    
    print("Final loss: {}".format(curr_mean_loss))
    print("Final accuracy: {}".format(mean_accuracy))
    
    return beta
    
```

```python
def plotbeta(beta):
    "The function which takes in beta and visualize it"
    beta = beta.reshape(dim_x, dim_y)
    plt.imshow(beta, cmap="gray")
    plt.title("Beta")
    plt.axis("off")
```

### 1.1 Effect of the number of batches:

Firstly, we use 1000 as the traning size and 200 as the test size. 

When the number of batches is 10, 30 and 50, the loss function converges at about 1000th iteration, 400th iteration and 200th iteration respectively. 

When the number of batches is 10, 30 and 50, the time per epoch is 5.11, 4.58 and 4.43 respectively.

Then, we use 2000 as the traning size and 400 as the test size.

When the number of batches is 10, 30 and 50, the loss function converges at about 2000th iteration, 1000th iteration and 600th iteration respectively. 

When the number of batches is 10, 30 and 50, the time per epoch is 12.58, 9.43 and 9.26 respectively.

When the number of batches is 10, 30 and 50, the final accuracy is 84.49%, 83.94% and 82.71% respectively.

Hence, in the following SGD on the whole sample, we would choose a relatively large number of batches to make the loss function converging fast, and without big sacrifices on accuracy and loss function.

We then compared step decay and 1/t decay to find the most efficient way of reducing the learning rate. 

Firstly, we used training set with size of 1000. We found that 1/t decay with lambda =0.015 may run faster and obtain lower loss than step decay did. Then, we increased the sample size to figure out how these two methods perform dealing with large data set. 

It turns out that the 1/t decay with lambda = 0.015 produced a better result, and thus we chose it for our final run.
